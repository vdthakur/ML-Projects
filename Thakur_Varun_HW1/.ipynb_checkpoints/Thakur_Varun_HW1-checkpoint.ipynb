## B(iii). Select the first 70 rows of Class 0 and the first 140 rows of Class 1 as the training set and the rest of the data as the test set.

```python
# create train and test sets
train_data = pd.concat([verterbral_dF[verterbral_dF['Classification'] == 0].iloc[0:70], verterbral_dF[verterbral_dF['Classification'] == 1].iloc[0:140]])
test_data = pd.concat([verterbral_dF[verterbral_dF['Classification'] == 0].iloc[70:], verterbral_dF[verterbral_dF['Classification'] == 1].iloc[140:]])

# adjusts training to have features and testing to have classes
X_train = train_data.drop('Classification', axis =1)
X_test = test_data.drop('Classification', axis =1)
y_train = train_data['Classification']
y_test = test_data['Classification']
```

## C(i). Write code for k-nearest neighbors with Euclidean metric

```python
train_errors = []
test_errors = []
k_axis_range = range(208, 0, -3)
for k in k_axis_range:
    knn = KNeighborsClassifier(n_neighbors=k, metric='euclidean')
    knn.fit(X_train, y_train)
    
    y_training_predictions = knn.predict(X_train.values)
    y_testing_predictions = knn.predict(X_test.values)
    
    train_accuracy = accuracy_score(y_train.values, y_training_predictions)
    test_accuracy = accuracy_score(y_test.values, y_testing_predictions)
    
    train_err = 1 - train_accuracy
    test_err = 1 - test_accuracy
    
    train_errors.append(train_err)
    test_errors.append(test_err)
    
```

## C(ii). Test all the data in the test database with k nearest neighbors. Take de- cisions by majority polling. Plot train and test errors in terms of k for k  {208, 205, . . . , 7, 4, 1, } (in reverse order). You are welcome to use smaller increments of k. Which k is the most suitable k among those values?

```python
plt.plot(k_axis_range, train_errors, label = "Train Error")
plt.plot(k_axis_range, test_errors, label = "Test Error")
plt.xlabel("Value of K")
plt.ylabel("Test Error")
plt.legend()
print("K Value with Smallest Test Error is K =",k_axis_range[np.argmin(test_errors)])
print("The Smallest Test Error is", round(min(test_errors),2))
```

K Value with Smallest Test Error is K = 4

The Smallest Test Error is 0.06

## C(ii). Confusion Matrix

```python
# from above can see that 4 is most optimal k
most_optimal_k = 4
knn_most_optimal = KNeighborsClassifier(n_neighbors=most_optimal_k, p=2, metric='euclidean')
knn_most_optimal.fit(X_train, y_train)
y_test_optimal_prediction = knn_most_optimal.predict(X_test.values)
most_optimal_confusion_matrix = confusion_matrix(y_test, y_test_optimal_prediction)
confusion_table = tabulate(most_optimal_confusion_matrix)
print("Confusion Matrix:")
print(most_optimal_confusion_matrix)
```

```python
# display confusion matrix has heatmap
print("Confusion Matrix as a Heatmap")
sns.heatmap(most_optimal_confusion_matrix, annot = True, xticklabels = ["Predicted Negative","Predicted Positive"], yticklabels = ["Actual Negative","Actual Positive"], cmap = "coolwarm")
print()
print("True Positive with index: [1,1] = 69")
print("True Negative with index: [0,0] = 25")
print("False Negative with index: [1,0] = 1")
print("False Positive with index: [0,1] = 5")
```

```python
# in [1,1] the first 1 represents actual positive, and 1 represents predicted positive
TP = most_optimal_confusion_matrix[1, 1]

# in [0,0] the first 0 represents actual neg, and 0 represents predicted neg
TN = most_optimal_confusion_matrix[0, 0]

# in [0,1] the first 0 represents actual negative, and 0 represents predicted positive
FP = most_optimal_confusion_matrix[0, 1]

# in [1,0] the first 1 represents actual positive, and 0 represents predicted negative
FN = most_optimal_confusion_matrix[1, 0]
```

## C(ii). True Positive Rate (Recall)

```python
true_positive_rate = (TP)/(TP+FN)
print("True Positive Rate =", round(true_positive_rate,2))
```

## C(ii). True Negative Rate

```python
true_negative_rate = (TN)/(TN+FP)
print("True Negative Rate =",round(true_negative_rate,2))
```

## C(ii). Precision

```python
precision = (TP)/(TP+FP)
print("Precision =",round(precision,2))
```

## C(ii). F1 Score

```python
f1_value = (2*(precision)*(true_positive_rate)/(precision + true_positive_rate))
print("F1 Score =",round(f1_value,2))
```

## C(iii). Since the computation time depends on the size of the training set, one may only use a subset of the training set. Plot the best test error rate which is obtained by some value of k, against the size of training set, when the size of training set is N  {10,20,30,...,210}.4 Note: for each N, select your training set by choosing the first N/3 rows of Class 0 and the first N  N/3 rows of Class 1 in the training set you created in 1(b)iii. Also, for each N, select the optimal k from a set starting from k = 1, increasing by 5. For example, if N = 200, the optimal k is selected from {1,6,11,...,196}. This plot is called a Learning Curve.


```python
error_list = []
N_list = []
# for every possible sample size
for n in range(10, 211, 10):
    best_test_error = float("inf")
    train_set = pd.concat([train_data[train_data["Classification"] == 0].iloc[0: n // 3], train_data[train_data["Classification"] == 1].iloc[0:n - (n // 3)]])
    X_train_new = train_set.drop('Classification', axis=1)
    y_train_new = train_set['Classification']
    
    # for every possible k based on sample size
    for k in range(1, n, 5):
        knn_new = KNeighborsClassifier(n_neighbors=k, metric='euclidean')
        knn_new.fit(X_train_new, y_train_new)
        
        y_test_predictions_new = knn_new.predict(X_test.values)
        
        test_accuracy_new = accuracy_score(y_test.values, y_test_predictions_new)
                            
        test_error_new = 1 - test_accuracy_new
        
        if test_error_new < best_test_error:
            best_test_error = test_error_new
    # append the sample size used and the best test error that was derived from the range of k based on that sample size       
    N_list.append(n)
    error_list.append(best_test_error)
```

### Learning Curve

```python
plt.plot(N_list, error_list,label = "Best Test Error Rate")
plt.title("Learning Curve")
plt.xlabel("N (Size of Training Set)")
plt.ylabel("Test Error")
plt.legend()
```

## D. Replace the Euclidean metric with the following metrics and test them. Sum- marize the test errors (i.e., when k = k) in a table. Use all of your training data and select the best k when {1,6,11,...,196}.

### (D)(i)(A). Minkowski Distance which becomes Manhattan Distance with p = 1.

```python
# empty dict to store k and it's test error
k_test_error_manhattan = {}
for k in range(1,200,5):
        knn_man = KNeighborsClassifier(n_neighbors=k, p=1, metric="manhattan")
        knn_man.fit(X_train, y_train)
        y_test_prediction = knn_man.predict(X_test.values)                 
        test_error_manhattan = 1 - accuracy_score(y_test.values, y_test_prediction)
        k_test_error_manhattan[k] = round(test_error_manhattan,2)
```

```python
min_error_manhattan = min(k_test_error_manhattan.values()) 
# store in list as there may be multiple
min_k_error_manhattan = [key for key, value in k_test_error_manhattan.items() if value == min_error_manhattan]
best_manhattan = {}   
for key in min_k_error_manhattan:    
    best_manhattan[key] = k_test_error_manhattan[key]
best_manhattan_df = pd.DataFrame(list(best_manhattan.items()), columns=['K', 'Test Error'])
best_manhattan_df
```

The most optimal K-Value for Manhattan Metric is k = 6 because it is the lowest number of k nearest neighbors in which the test error is the lowest

The best test error for Manhattan Metric is 0.11

### (D)(i)(B). Minkowski Distance with log10(p)  {0.1,0.2,0.3,...,1}. In this case, use the k you found for the Manhattan distance in 1(d)iA. What is the best log10(p)?.

```python
# create a list to store test errors
k_test_error_minkowski = []
log_p = np.logspace(0.1,1,10)
# k found and used from above in Manhattan calculation
most_optimal_k_manhattan = 6
for p in log_p:
        knn_mink = KNeighborsClassifier(n_neighbors=most_optimal_k_manhattan,p = p, metric="minkowski")
        knn_mink.fit(X_train, y_train)
        y_test_prediction = knn_mink.predict(X_test.values)                 
        test_error_minkowski = 1 - accuracy_score(y_test.values, y_test_prediction)
        
        # find test error for every p and add to list
        k_test_error_minkowski.append(round(test_error_minkowski,2))
```

```python
col_names = ["p","log10(p)","Test Error"]
possible_p_values = list(np.linspace(0.1, 1, 10))
log_10_p = list(log_p)
minkowski_df = pd.DataFrame({"p": possible_p_values,"log10(p)": log_10_p,"Test Error": k_test_error_minkowski})
```

```python
min(k_test_error_minkowski)
```

```python
# checking to see if the minimum test error appears more than once
k_test_error_minkowski.index(0.06)
```

```python
# display row with best test error
minkowski_df[minkowski_df['Test Error'] == min(k_test_error_minkowski)]
```

The optimal P-Value for Minkowski is 0.6

The best test error for Minkowski is 0.06

### (D)(i)(C). Minkowski Distance which becomes Chebyshev Distance with p  

```python
k_test_error_chebyshev = {}
for k in range(1,200,5):
    knn_cheb = KNeighborsClassifier(n_neighbors=k,p=float("inf"), metric="chebyshev")
    knn_cheb.fit(X_train, y_train)
    y_test_prediction = knn_cheb.predict(X_test.values)                 
    test_error_chebyshev = 1 - accuracy_score(y_test.values, y_test_prediction)
    k_test_error_chebyshev[k]=round(test_error_chebyshev,2)
```

```python
min_error_chebyshev = min(k_test_error_chebyshev.values()) 
# store in a list as there may be multiple
min_k_error_chebyshev = [key for key, value in k_test_error_chebyshev.items() if value == min_error_chebyshev]
# store the optimal K's with their test errors
best_chebyshev = {}   
for key in min_k_error_chebyshev:    
    best_chebyshev[key] = k_test_error_chebyshev[key]
best_chebyshev_df = pd.DataFrame(list(best_chebyshev.items()), columns=['K', 'Test Error'])
best_chebyshev_df
```

The optimal K-value for Chebyshev is k = 16

The best test error for Chebyshev is 0.08

### (D)(ii.) Mahalanobis Distance. Mahalanobis Distance requires inverting the covariance matrix of the data. When the covariance matrix is singular or ill-conditioned, the data live in a linear subspace of the feature space. In this case, the features have to be transformed into a reduced feature set in the linear subspace, which is equivalent to using a pseudoinverse instead of an inverse.


```python
covariance_matrix = X_train.cov()
inverse_covariance_matrix = np.linalg.inv(covariance_matrix)
inverse_covariance_matrix.shape
```

```python
k_test_error_mahalanobis = {}
for k in range(1,200,5):
    knn_mahala = KNeighborsClassifier(n_neighbors=k,metric="mahalanobis",metric_params={'VI':inverse_covariance_matrix})
    knn_mahala.fit(X_train, y_train)
    y_test_prediction = knn_mahala.predict(X_test.values)                 
    test_error_mahalanobis = 1 - accuracy_score(y_test.values, y_test_prediction)
    k_test_error_mahalanobis[k]=round(test_error_mahalanobis,2)
```

```python
# find the value of best test error
min_error_mahalanobis = min(k_test_error_mahalanobis.values()) 
min_k_error_mahalanobis = [key for key, value in k_test_error_mahalanobis.items() if value == min_error_mahalanobis]
best_mahalanobis = {}   
for key in min_k_error_mahalanobis:    
    best_mahalanobis[key] = k_test_error_mahalanobis[key]
best_mahalanobis_df = pd.DataFrame(list(best_mahalanobis.items()), columns=['K', 'Test Error'])
best_mahalanobis_df
```

The most optimal K-Value for Mahalanobis Metric is k = 6 because k = 1 can result in overfitting.

### Summarized Test Errors from Part D

## (E). The majority polling decision can be replaced by weighted decision, in which the weight of each point in voting is inversely proportional to its distance from the query/test data point. In this case, closer neighbors of a query point will have a greater influence than neighbors which are further away. Use weighted voting with Euclidean, Manhattan, and Chebyshev distances and report the best test errors when k  {1,6,11,16,...,196}.

```python
measurement_metric = ['euclidean', 'manhattan', 'chebyshev']

euclidean_k_error = {}
manhattan_k_error = {}
chebyshev_k_error = {}

for metric in measurement_metric:
    for k in range(1,200,5):
        knn_weighted = KNeighborsClassifier(n_neighbors=k, weights='distance', metric=metric)
        knn_weighted.fit(X_train, y_train)
        y_test_prediction = knn_weighted.predict(X_test.values)

        test_error = 1 - accuracy_score(y_test.values, y_test_prediction)
        
        if metric == measurement_metric[0]:
            euclidean_k_error[k] = test_error
        elif metric == measurement_metric[1]:
            manhattan_k_error[k] = test_error
        elif metric == measurement_metric[2]:
            chebyshev_k_error[k] = test_error
```

```python
best_euclidean_weighted = {}   
min_k_error_euclidean = [key for key, value in euclidean_k_error.items() if value == min(euclidean_k_error.values())]
for key in min_k_error_euclidean:    
    best_euclidean_weighted[key] = euclidean_k_error[key]
best_euclidean_weighted_df = pd.DataFrame(list(best_euclidean_weighted.items()), columns=['K', 'Test Error'])
best_euclidean_weighted_df
```

The Optimal K-Value for Weighted Euclidean is k = 6

The Best Test Error for Weighted Euclidean is 0.10

```python
best_manhattan_weighted = {}   
min_k_error_manhattan = [key for key, value in manhattan_k_error.items() if value == min(manhattan_k_error.values())]
for key in min_k_error_manhattan:    
    best_manhattan_weighted[key] = manhattan_k_error[key]
best_manhattan_weighted_df = pd.DataFrame(list(best_manhattan_weighted.items()), columns=['K', 'Test Error'])
best_manhattan_weighted_df
```

The Optimal K-Value for Weighted Manhattan is k = 26

The Best Test Error for Weighted Manhattan is 0.10

```python
best_chebyshev_weighted = {}   
min_k_error_chebyshev = [key for key, value in chebyshev_k_error.items() if value == min(chebyshev_k_error.values())]
for key in min_k_error_chebyshev:    
    best_chebyshev_weighted[key] = chebyshev_k_error[key]
best_chebyshev_weighted_df = pd.DataFrame(list(best_chebyshev_weighted.items()), columns=['K', 'Test Error'])
best_chebyshev_weighted_df
```

The Optimal K-Value for Weighted Chebyshev is k = 16 as it provides the lowest test error with the smallest number of nearest neighbors

The Best Test Error for Weighted Chebyshev is 0.11

### Summarize Test Errors from Part E (Weighted Metrics)

```python
weighted_manhattan_table = tabulate(best_euclidean_weighted_df, headers = 'keys', tablefmt = "fancy_grid", showindex=False)
weighted_euclidean_table = tabulate(best_manhattan_weighted_df, headers = 'keys', tablefmt = "fancy_grid", showindex=False)
weighted_chebyshev_table = tabulate(best_chebyshev_weighted_df, headers = 'keys', tablefmt = "fancy_grid", showindex=False)



print("The Best Test Errors and Optimal K-Values for WEIGHTED Mahalobis Metric is:")
print(weighted_manhattan_table)
print()
print("The Best Test Errors and Optimal K-Values for WEIGHTED Euclidean Metric is:")
print(weighted_euclidean_table)
print()
print("The Best Test Errors and Optimal K-Values for WEIGHTED Chebyshev Metric are:")
print(weighted_chebyshev_table)
```
